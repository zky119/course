### 需求：不同年月份 销售额趋势分析

```python
# 取出年月作为新列
df1['date1'] = pd.to_datetime(df1['创建时间']).apply(lambda x: x.strftime('%Y-%m'))
df1['date1']
```

```python
df2 = df1.groupby('date1')['date1','实付金额'].sum()
```

```python
plt.figure(figsize=(20,8),dpi = 80)
plt.xticks(rotation = 45)
plt.plot(df2.index,df2['实付金额'],linestyle = '--',linewidth=4,color = 'lightblue',alpha = 1,marker = '>')
plt.savefig('./a.png')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(range(10))
ax.set_xlabel('company')
ax.set_ylabel('gmv')
ax.spines[['bottom','left','right','top']].set_color('green')
ax.xaxis.label.set_color('blue')
ax.yaxis.label.set_color('blue')
ax.tick_params(axis='x',color = 'red')
plt.show()
```

### 设置坐标轴范围

```python
plt.rcParams["axes.unicode_minus"]=False
```

```python
x = np.arange(-10,11,1)
y = x**2
plt.plot(x,y)
plt.xlim(xmin = 0)   # 限制 x 轴 坐标
plt.ylim(ymin = 0)   # 限制 y 轴 坐标
plt.show()
```

### 更改坐标轴方式

```python
y = range(0,14,2)
x = [-3,-2,-1,0,1,2,3]
ax = plt.gca()

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_color('blue')
ax.spines['left'].set_color('red')

ax.spines['bottom'].set_position(('data',6))
ax.spines['left'].set_position(('data',0))
plt.plot(x,y)
plt.show()
```
