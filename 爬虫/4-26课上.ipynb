### 取影评地址

```python
base_url = 'https://movie.douban.com/subject/1291583/reviews?start=0'
ua = UserAgent()
headers = {'User-Agent':ua.chrome}
response = requests.get(base_url,headers=headers)
# response.text
root = etree.HTML(response.text)
hrefs = root.xpath('//div[@class="main-bd"]/h2/a/@href')
print(hrefs)
```

### 取影评内容

```python
for url in hrefs:
    # print(url)
    resp = requests.get(url,headers = headers)
    root2 = etree.HTML(resp.text)
    content = root2.xpath('//div[@class="review-content clearfix"]/text()')
    # print(content)
    s = ''
    if len(content) == 2:
        content = root2.xpath('//div[@class="review-content clearfix"]/p/text()')
        for i in content:
            s += i.replace('\n','').strip()
    else:
        for i in content:
            s += i.replace('\n','').strip()
    print(s)
```

### 多页取影评

```python
for i in range(0,1293,20):
    base_url = f'https://movie.douban.com/subject/1291583/reviews?start={i}'
    ua = UserAgent()
    headers = {'User-Agent':ua.chrome}
    response = requests.get(base_url,headers=headers)
    # response.text
    root = etree.HTML(response.text)
    hrefs = root.xpath('//div[@class="main-bd"]/h2/a/@href')
    for url in hrefs:
        # print(url)
        resp = requests.get(url,headers = headers)
        root2 = etree.HTML(resp.text)
        content = root2.xpath('//div[@class="review-content clearfix"]/text()')
        # print(content)
        s = ''
        if len(content) == 2:
            content = root2.xpath('//div[@class="review-content clearfix"]/p/text()')
            for i in content:
                s += i.replace('\n','').strip()
        else:
            for i in content:
                s += i.replace('\n','').strip()
        # print(s)
        with open('./pinglun2.csv',mode='a',newline='',encoding='utf-8') as f:
            csvfile = csv.writer(f)
            csvfile.writerow([s])
        time.sleep(random.uniform(1,3))
print('wancheng')
```

```python
import pandas as pd
df = pd.read_csv('./pinglun.csv',header=None)
list_ = df[0].tolist()
```

```python
import jieba.analyse
dict_ = {}
jieba.del_word('天空')
jieba.del_word('之城')
jieba.add_word('天空之城')

for e in list_:
    cis = jieba.analyse.extract_tags(str(e))
    for e2 in cis:
        try:
            dict_[e2] += 1
        except:
            dict_[e2] = 1
dict_
```

```python
data = []
for k,v in dict_.items():
    if len(k) >= 2:
        data.append((k,v))
data
```

```python
c=(
    WordCloud(init_opts=opts.InitOpts())
    .add(series_name="热点分析", data_pair=data, word_size_range=[10, 100])
    .set_global_opts(
        title_opts=opts.TitleOpts(
        title="热点分析",
        title_textstyle_opts=opts.TextStyleOpts(font_size=23)
    ),
    tooltip_opts=opts.TooltipOpts(is_show=True),
    )
    )
c.render_notebook()
```
