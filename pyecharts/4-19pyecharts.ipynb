### 折线图

```python
import pandas as pd 
data1 = pd.read_excel('./超市.xlsx')
data1.head()
```

```python
data1.columns
```

```python
c = (
    Line() # 创建折线图对象
    .add_xaxis(Faker.choose()) # 添加X轴数据
    .add_yaxis("商家A", Faker.values()) # 添加Y轴数据
    .add_yaxis("商家B", Faker.values()) # 添加Y轴数据
    .set_global_opts(title_opts = opts.TitleOpts(title="Line-基本⽰例")) 
)
c.render_notebook()
```

```python
data1['下单年月_data'] = data1['订单日期'].dt.strftime('%Y-%m')
data1
```

### 需求：按年月看销售额趋势分析

```python
data2 = data1.groupby('下单年月_data')['销售额'].sum()
data2
```

### 可视化

```python
x = data2.index.tolist()
c = (
    Line() # 创建折线图对象
    .add_xaxis(data2.index.tolist()) # 添加X轴数据
    .add_yaxis("总销售额", data2) # 添加Y轴数据
    .set_global_opts(
        # 设置 图例 显示 False
        legend_opts=opts.LegendOpts(is_show=False),
        # 设置 标题 和 标题居中
        title_opts = opts.TitleOpts(title="销售额趋势",pos_left='center'),
        # 设置x轴名称 和 大小
        xaxis_opts=opts.AxisOpts(name='日期',axislabel_opts={"rotate":45},name_textstyle_opts=opts.TextStyleOpts(font_size=20)),
        # 设置y轴名称 和 大小
        yaxis_opts=opts.AxisOpts(name='销售额',name_textstyle_opts=opts.TextStyleOpts(font_size=20))
        ) 
)
c.render_notebook()
```

### 面积图 原生

#### 需求：销售额和利润月度走势

```python
data3 = data1.groupby('下单年月_data').agg({'销售额':'sum','利润':'sum'})
data3
```

```python
x = data3.index.tolist()
y1 = data3['销售额'].tolist()
y2 = data3['利润'].tolist()
```

```python
c = (
    Line()
    .add_xaxis(x) 
    # 添加Y1轴数据
    .add_yaxis(
        "总销售额",
        y1,
        areastyle_opts = opts.AreaStyleOpts(opacity = 0.5,color='#C7ECEE'),
        is_smooth = True
        ) 
    # 添加Y2轴数据
    .add_yaxis(
        "总销售额", 
        y2,
        areastyle_opts = opts.AreaStyleOpts(opacity = 0.5),
        is_smooth = True
        ) 
    .set_global_opts(
        # 设置 图例 显示 False
        legend_opts=opts.LegendOpts(is_show=False),
        # 设置 标题 和 标题居中
        title_opts = opts.TitleOpts(title="销售额趋势",pos_left='center'),
        # 设置x轴名称 和 大小
        xaxis_opts=opts.AxisOpts(name='日期',axislabel_opts={"rotate":45},name_textstyle_opts=opts.TextStyleOpts(font_size=20)),
        # 设置y轴名称 和 大小
        yaxis_opts=opts.AxisOpts(name='价格',name_textstyle_opts=opts.TextStyleOpts(font_size=20))
        ) 
)
c.render_notebook()
```

```python
c = (
    Line() # 创建折线图对象
    .add_xaxis(x) 
    # 添加Y1轴数据
    .add_yaxis(
        "总销售额",
        y1,
        areastyle_opts = opts.AreaStyleOpts(opacity = 0.5,color='#F7DC6F'),
        is_smooth = True
        ) 
    # 添加Y2轴数据
    .add_yaxis(
        "利润", 
        y2,
        areastyle_opts = opts.AreaStyleOpts(opacity = 0.5),
        is_smooth = True
        ) 
    .set_global_opts(
        # 设置 图例 显示 False
        legend_opts=opts.LegendOpts(pos_top='5%', pos_right='5%'),
        # 设置 标题 和 标题居中
        title_opts = opts.TitleOpts(title="销售额趋势",pos_left='center'),
        # 设置x轴名称 和 大小
        xaxis_opts=opts.AxisOpts(name='日期',axislabel_opts={"rotate":45},name_textstyle_opts=opts.TextStyleOpts(font_size=20)),
        # 设置y轴名称 和 大小
        yaxis_opts=opts.AxisOpts(name='价格',name_textstyle_opts=opts.TextStyleOpts(font_size=20))
        ) 
    .set_series_opts(
        # 不显示数值标签
        label_opts=opts.LabelOpts(is_show=False)
        )
)
c.render_notebook()
```

##### 原生堆叠

```python
from pyecharts.charts import Bar
from pyecharts.faker import Faker


c = (
    Bar()
    .add_xaxis(Faker.choose())
    .add_yaxis("商家A", Faker.values(), stack="stack1")
    .add_yaxis("商家B", Faker.values(), stack="stack1")
    .add_yaxis("商家C", Faker.values())
    .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
    .set_global_opts(title_opts=opts.TitleOpts(title="Bar-堆叠数据（部分）"))
)
c.render_notebook()
```

```python
data5 = data1.groupby(['地区','类别'])['销售额'].sum()
data5 = pd.DataFrame(data5).unstack()
data5
```

```python
x = data5.index.tolist()
y1 = data5['销售额']['办公用品'].tolist()
y2 = data5['销售额']['家具'].tolist()
y3 = data5['销售额']['技术'].tolist()
```

```python
from pyecharts.charts import Bar
c = (
    Bar()
    .add_xaxis(x)
    .add_yaxis("办公用品", y1, stack="stack1")  # y1  stack="stack1" stack1 是让这仨y堆到 这叫stack1 的柱子上
    .add_yaxis("家具", y2, stack="stack1")   # y2
    .add_yaxis("技术", y3, stack="stack1")   # y3
    .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
    .set_global_opts(title_opts=opts.TitleOpts(title="Bar-堆叠数据（部分）"))
)
c.render_notebook()
```

### 饼图

##### 需求：不同类别销售额的占比分析

```python
datapie = data1.groupby('类别')['销售额'].sum()
# 为了转成老师的格式
datapie = pd.DataFrame(datapie)
datapie
```

```python
lei = datapie.index.tolist()
sale = datapie['销售额'].tolist()
```

```python
from pyecharts.charts import Pie

c = (
    Pie()
    .add("", [list(z) for z in zip(lei, sale)])
    .set_global_opts(title_opts=opts.TitleOpts(title="不同类别销售额的占比分析"))
    .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c}"))
)
c.render_notebook()
```

#### 圆环图

```python
from pyecharts.charts import Pie

c = (
    Pie()                                     # ↓  radius：半径
    .add("", [list(z) for z in zip(lei, sale)],radius=['45%','55%'])
    .set_global_opts(title_opts=opts.TitleOpts(title="不同类别销售额的占比分析"))
    .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c}"))
)
c.render_notebook()
```

#### 南丁格尔玫瑰图

```python
from pyecharts.charts import Pie

c = (
    Pie()                                     # ↓ rosetype：玫瑰类型  ↓  radius：半径 
    .add("", [list(z) for z in zip(lei, sale)],rosetype='radius',radius='55%')
    .set_global_opts(title_opts=opts.TitleOpts(title="不同类别销售额的占比分析"))
    .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c}"))
)
c.render_notebook()
```

#### 矩形树图

##### 需求：不同省份销售额对比分析

```python
data1.columns
```

```python
data7 = data1.groupby('省/自治区')['销售额'].sum()
# 为了迎合 老师的 DataFrame
data7 = pd.DataFrame(data7)
data7.head()
```

```python
# 获得数据
tree_data = []  # 初始化 一个列表
sample = {}   # 初始化一个字典
# 遍历一遍改成字典 再 添加到列表中
for i,j in zip(data7.index,data7['销售额']):
    sample = {'value':j,'name':i}
    tree_data.append(sample)
```

```python
# 可视化
# 导入 树图
from pyecharts.charts import TreeMap

c = (
    TreeMap()
    .add("", tree_data)
    .set_global_opts(
        title_opts=opts.TitleOpts(title="Tree-基本示例"),
        visualmap_opts=opts.VisualMapOpts(min_=10000,max_=1600000,range_color=["#cf857f","#a81a0f"],pos_bottom='10%')
        )
)
c.render_notebook()
```

### 散点图

##### 需求：分析各子类别销售额和利润之间的关系

```python
data8 = data1.groupby('子类别')['销售额','利润'].sum()
data8.head()
```

```python
# 导入 散点图
from pyecharts.charts import Scatter
# 导入 jscode
from pyecharts.commons.utils import JsCode

# 给 x，y 数据 进行赋值
x_data = data8['销售额'].tolist()
y_data = data8['利润'].tolist()
z_data = data8.index.tolist()
# 下面是复制官网的，不用动
c = (
    Scatter()
    .add_xaxis(xaxis_data=x_data)
    .add_yaxis(
        "商家A",
        [list(z) for z in zip(y_data, z_data)],
        label_opts=opts.LabelOpts(
            formatter=JsCode(
                "function(params){return  params.value[2]}"
            ),
            position='right'   # 设置这个，让文本在右边显示
        ),
    )
    .set_series_opts(

    )
    .set_global_opts(
        xaxis_opts=opts.AxisOpts(
            # splitline_opts 这个是刻度线 网格 x轴的
            type_="value", splitline_opts=opts.SplitLineOpts(is_show=True)
        ),
        yaxis_opts=opts.AxisOpts(
            type_="value",
            axistick_opts=opts.AxisTickOpts(is_show=True),
            # splitline_opts 这个是刻度线 网格 y轴的
            splitline_opts=opts.SplitLineOpts(is_show=True),
        ),
        tooltip_opts=opts.TooltipOpts(
            formatter=JsCode(
                "function (params) {return  params.value[2] + ':' + params.value[1]}"
            )
        ),
        visualmap_opts=opts.VisualMapOpts(
            type_="color", max_=150, min_=20, dimension=1
        ),

)
)
c.render_notebook()
```
