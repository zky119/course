```python
import numpy as np
import pandas as pd
```

```python
order_data = pd.read_excel('超市.xlsx')
order_data.head()
```

```python
provinceincome_tab = order_data.groupby(['省/自治区'])['销售额'].sum()
provinceincome_tab = pd.DataFrame(provinceincome_tab)
provinceincome_tab.head(1)
```

```python
x_data = provinceincome_tab.index.to_list()
y_data = provinceincome_tab['销售额'].tolist ()
[list(z) for z in zip(y_data, x_data)]
```

```python
import pyecharts.options as opts
from pyecharts.charts import Map

userprovince_chart = (
    Map()
    .add(
        'geo',
        [list(z) for z in zip(y_data, x_data)] ,
        "china"
        )
    .set_global_opts(
        visualmap_opts=opts.VisualMapOpts(
            min_=provinceincome_tab.iloc[:,-1].min(),
            max_=provinceincome_tab.iloc[:,-1].max(),
            range_color=['lightyellow','orangered'],
            precision=2
            ),
        title_opts=opts.TitleOpts(
            '各省市销售额热力地图',
            pos_left='center'
            ),
        legend_opts=opts.LegendOpts(is_show=False)
        )
    .set_series_opts(
        tooltip_opts=opts.TooltipOpts(formatter="{b}:{c}元")
        )
    )
userprovince_chart.render_notebook()
```
