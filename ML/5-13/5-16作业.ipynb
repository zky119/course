```python
import pandas as pd 
import numpy as np 
```

```python
class SGD:
    def __init__(self, eta=0.1, epoch=10):
        self.eta = eta 
        self.epoch = epoch

    def fit(self, X, y):
        one = np.ones((X.shape[0],1))
        X = np.concatenate([one,X], axis=1)
        self.w = np.random.random(X.shape[1])
        for _ in range(self.epoch):
            order = np.random.permutation(len(X))
            for xi, yi in zip(X[order],y[order]):
                y_hat = self.w @ xi
                self.w += self.eta * (yi - y_hat) * xi
        return self
    def predict(self, X):
        return X @ self.w
```

```python
from sklearn.datasets import make_regression
X, y, coef = make_regression(n_samples=1000, n_features=5, bias=2.5, coef=True,noise=5, random_state=0)
print(f"真实权重：{coef}")

```

```python
sgd = SGD()
sgd.fit(X,y)
print(sgd.w)
```
